```python
# When were the reservoirs discovered in Venezuela?
con.execute("""
SELECT
    Year_Discovered,
    COUNT(*) AS reservoirs_discovered
FROM reservoirs
GROUP BY Year_Discovered
ORDER BY Year_Discovered;
""").df()
```

```python
#Which location holds the most oil reserves?
con.execute("""
SELECT
    Location,
    SUM(Proven_Reserves_Billion_Barrels) AS total_proven_reserves
FROM reservoirs
GROUP BY Location
ORDER BY total_proven_reserves DESC;
""").df()
```

```python
con.execute("""
CREATE OR REPLACE TABLE reservoirs_clean AS
SELECT
    Location AS location,
    Year_Discovered AS year_discovered,

    /* Proven Reserves (already numeric OR text-safe) */
    CAST(Proven_Reserves_Billion_Barrels AS DOUBLE)
        AS proven_reserves_bbb,

    /* Estimated Recoverable Reserves (ranges handled safely) */
    CASE
        WHEN CAST(Estimated_Recoverable_Reserves_Billion_Barrels AS VARCHAR) LIKE '%-%' THEN
            (
                CAST(
                    SPLIT_PART(
                        CAST(Estimated_Recoverable_Reserves_Billion_Barrels AS VARCHAR),
                        '-', 1
                    ) AS DOUBLE
                )
                +
                CAST(
                    SPLIT_PART(
                        CAST(Estimated_Recoverable_Reserves_Billion_Barrels AS VARCHAR),
                        '-', 2
                    ) AS DOUBLE
                )
            ) / 2
        ELSE
            CAST(Estimated_Recoverable_Reserves_Billion_Barrels AS DOUBLE)
    END AS recoverable_reserves_bbb,

    /* Production Capacity */
    CAST(Production_Capacity_Barrels_Day AS BIGINT)
        AS production_capacity_bpd

FROM reservoirs
WHERE
    Proven_Reserves_Billion_Barrels IS NOT NULL
    AND Estimated_Recoverable_Reserves_Billion_Barrels IS NOT NULL;
""")
```

```python
con.execute("""
SELECT
    Location,
    ROUND(
        SUM(Estimated_Recoverable_Reserves_Billion_Barrels)
        / NULLIF(SUM(Proven_Reserves_Billion_Barrels), 0),
        2
    ) AS recoverability_ratio
FROM reservoirs
GROUP BY Location
ORDER BY recoverability_ratio DESC;
""").df()
```

```python
# Are high -reserve fields producing aggresively?
con.execute("""
SELECT
    Location,
    SUM(Production_Capacity_Barrels_Day) AS total_production_bpd,
    SUM(Proven_Reserves_Billion_Barrels) AS proven_reserves
FROM reservoirs
GROUP BY Location
ORDER BY total_production_bpd DESC;
""").df()
```

```python
# Which reservoirs matter most overall?
con.execute("""
SELECT
    location,
    year_discovered,
    proven_reserves_bbb,
    production_capacity_bpd,
    recoverable_reserves_bbb,
    (
        proven_reserves_bbb * 0.4 +
        recoverable_reserves_bbb * 0.4 +
        production_capacity_bpd / 100000.0 * 0.2
    ) AS strategic_score
FROM reservoirs_clean
ORDER BY strategic_score DESC
LIMIT 10;
""").df()
```

```python
# Production vs Reserves Analysis:
con.execute("""
SELECT
    location,
    SUM(production_capacity_bpd) AS production_bpd,
    SUM(proven_reserves_bbb) AS reserves_bbb
FROM reservoirs_clean
GROUP BY location
ORDER BY production_bpd DESC;
""").df()
```

```python
con.execute("""
DROP TABLE IF EXISTS reservoirs_clean;

CREATE TABLE reservoirs_clean AS
SELECT
    Location AS location,

    /* Handle year values like '1970s', '1980s', numeric years */
    CASE
        WHEN CAST(Year_Discovered AS VARCHAR) LIKE '%s' THEN
            CAST(
                SUBSTR(CAST(Year_Discovered AS VARCHAR), 1, 4)
                AS INTEGER
            ) + 5
        ELSE
            CAST(Year_Discovered AS INTEGER)
    END AS year_discovered,

    /* Proven reserves */
    CAST(
        REPLACE(CAST(Proven_Reserves_Billion_Barrels AS VARCHAR), ',', '')
        AS DOUBLE
    ) AS proven_reserves_bbb,

    /* Estimated recoverable reserves (ranges handled) */
    CASE
        WHEN CAST(Estimated_Recoverable_Reserves_Billion_Barrels AS VARCHAR) LIKE '%-%' THEN
            (
                CAST(
                    SPLIT_PART(
                        REPLACE(CAST(Estimated_Recoverable_Reserves_Billion_Barrels AS VARCHAR), ',', ''),
                        '-', 1
                    ) AS DOUBLE
                )
                +
                CAST(
                    SPLIT_PART(
                        REPLACE(CAST(Estimated_Recoverable_Reserves_Billion_Barrels AS VARCHAR), ',', ''),
                        '-', 2
                    ) AS DOUBLE
                )
            ) / 2
        ELSE
            CAST(
                REPLACE(CAST(Estimated_Recoverable_Reserves_Billion_Barrels AS VARCHAR), ',', '')
                AS DOUBLE
            )
    END AS recoverable_reserves_bbb,

    /* Production capacity */
    CAST(
        REPLACE(CAST(Production_Capacity_Barrels_Day AS VARCHAR), ',', '')
        AS BIGINT
    ) AS production_capacity_bpd

FROM reservoirs
WHERE
    Proven_Reserves_Billion_Barrels IS NOT NULL
    AND Estimated_Recoverable_Reserves_Billion_Barrels IS NOT NULL
    AND Production_Capacity_Barrels_Day IS NOT NULL;
""")
```

```python
con.execute("DESCRIBE reservoirs_clean").df()
```

```python
#Strat Reservoir Ranking:
con.execute("""
SELECT
    location,
    year_discovered,
    proven_reserves_bbb,
    recoverable_reserves_bbb,
    production_capacity_bpd,
    (
        proven_reserves_bbb * 0.4 +
        recoverable_reserves_bbb * 0.4 +
        production_capacity_bpd / 100000.0 * 0.2
    ) AS strategic_score
FROM reservoirs_clean
ORDER BY strategic_score DESC
LIMIT 10;
""").df()
```

```python
con.execute("""
SELECT
    COUNT(*) AS total_rows,
    COUNT(DISTINCT location) AS unique_locations,
    SUM(CASE WHEN proven_reserves_bbb IS NULL THEN 1 ELSE 0 END) AS missing_proven,
    SUM(CASE WHEN recoverable_reserves_bbb IS NULL THEN 1 ELSE 0 END) AS missing_recoverable
FROM reservoirs_clean;
""").df()
```
